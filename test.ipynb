```python
def softmax(Z):
    S=0
    for z in Z:
        S+=np.exp(z)
    return np.exp(Z)/S
```

```python
def jacobian_soft(S):
    J=[]
    for i in range(len(S)):
        row=[]
        for k in range(len(S)):
            if k==i:
                row.append(S[k]-S[k]**2)
            else :
                row.append(-S[k]*S[i])
        J.append(row)
    return np.array(J)
```
